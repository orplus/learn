```python
my_url = 'https://www.theshop.mu/collections/breakfast-1'

#opening up connection, grabbing the page
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
page_soup = soup(page_html, "html.parser")
print(page_soup)
```

```python
#to get all links
for link in page_soup.find_all('a'):
    print(link.get('href'))
```

```python
#grabs each product
containers = page_soup.findAll("div", {"class":"productitem"})
print(containers)
```

```python
container = containers[0]
print(container)
```

# Loop to output all products in dataframe

```python
data = {
    'Category': [], 'Product Name': [], 'Price': [], 'Image': []
}
for container in containers:
    
    #category
    data['Category'].append(page_soup.h1.text.strip())
    
    #supplier_name, product_name, weight
    product_name = container.div.h2.a.text.strip()
    product = product_name.replace(",", " | ") 
    a = product.strip('|')
    data['Product Name'].append(a)

    #price of product
    price_container = container.findAll("span", {"class": "money"})
    price = price_container[0].text.strip()
    data['Price'].append(price)
    
    #image
    data['Image'].append(container.img["data-src"])
    
    df = pd.DataFrame(data)
```

```python
urls = ['https://www.theshop.mu/collections/fresh-products',
        'https://www.theshop.mu/collections/frozen', 
        'https://www.theshop.mu/collections/pantry', 
        'https://www.theshop.mu/collections/breakfast-1', 
        'https://www.theshop.mu/collections/beverages-1']

for url in urls:
    uClient = uReq(url)
    page_html = uClient.read()
    #sleep(randint(2,10))
    #uClient.close()
    page_soup = soup(page_html, "html.parser")
    #print(page_soup)
    containers = page_soup.findAll("div", {"class":"productitem"})
    #print(containers)
    containers[0]
    container = containers[0]
    data = {
    'Category': [], 'Product Name': [], 'Price': [], 'Image': []
    }
    for container in containers:

        #category
        data['Category'].append(page_soup.h1.text.strip())

        #supplier_name, product_name, weight
        product_name = container.div.h2.a.text.strip()
        #product = product_name.replace(",", " | ") 
        #a = product.strip('|')
        data['Product Name'].append(product_name)

        #price of product
        price_container = container.findAll("span", {"class": "money"})
        price = price_container[0].text.strip()
        data['Price'].append(price)

        #image
        data['Image'].append(container.img["data-src"])

df = pd.DataFrame(data)
```
